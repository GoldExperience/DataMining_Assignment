## one sentence example

```
#@title Randomly get a comment
import random
i=int(random.random()*len(reviews))

selected = reviews.iloc[i]
text = selected["comments"]
text_id = selected["id"]
print(text_id)
print(text)
```

```
#@title remove punctuation and "\n"
import string
text = text.translate(str.maketrans('','',string.punctuation)).replace('\n',"")
print(text)
```

```
#@title remove stop words
from nltk import word_tokenize
from nltk.corpus import stopwords
a = set(stopwords.words('english'))
text_lower = word_tokenize(text.lower())
text_remove_stopwords = [x for x in text_lower if x not in a]
```

```
#@title Stemming
from nltk.stem import PorterStemmer
pst = PorterStemmer()
text_stemed = []
text_stemed = [pst.stem(x) for x in text_remove_stopwords]
```

```
#@title Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 
text_lemmatization = [lemmatizer.lemmatize(x) for x in text_remove_stopwords]
```

```
#@title Part of speech tagging
text_pos_tag = [nltk.pos_tag([x]) for x in text_remove_stopwords]
pos_tag=[]
for word in text_pos_tag:
  pos_tag.append(word[0][1])
```

```
#@title put together as dataframe
ziplist = list(zip(text_remove_stopwords,text_stemed,text_lemmatization,pos_tag))
df = pd.DataFrame(ziplist,columns=["removed_stopwords","stemed","lemmatization","pos_tag"])
```
